```python
import pandas as pd
import matplotlib.pyplot as plt

#load CSV
df = pd.read_csv("./IT_Employees_Info_2000.csv")

# Get department counts
counts = df['Department'].value_counts()

# Create explode list (highlight the first slice)
explode = [0.1] + [0] * (len(counts) - 1)

#plot pie chart
plt.figure(figsize=(5, 4))
plt.pie(counts, labels=counts.index, autopct='%1.1f%%', shadow=True, explode=explode)

plt.title('Pie Chart of Department Distribution')
plt.tight_layout()
plt.show()
```
